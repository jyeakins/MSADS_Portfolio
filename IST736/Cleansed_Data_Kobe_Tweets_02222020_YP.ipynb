```
#Cleansing the data
def clean_rogue_characters(string):
    exclude = ['\\',"\'",'"']
    string = ''.join(string.split('\\n'))
    string = ''.join(ch for ch in string if ch not in exclude)
    return string

df['original_text'] = df['original_text'].apply( lambda x: clean_rogue_characters(x) )
df['original_text'][0]
```

```
df = pd.read_csv('cleansed_data.csv')
df[:10]
```

```
#Remove all Twitter handles and hashtags
df['original_text'] = df['original_text'].astype(str)
for x in range(0, len(df)): 
  df.iloc[x,0]= re.sub(r'@[A-Za-z]*', '', df.iloc[x,0], flags=re.IGNORECASE) 
  df.iloc[x,0]= re.sub(r'#[A-Za-z]*', '', df.iloc[x,0], flags=re.IGNORECASE)
```

```
#Convert to lowercase
df['original_text'] = df['original_text'].apply(lambda x: x.lower())
```

```
#Strip punctuation
df['original_text'] = df['original_text'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))
```
